# LECTURA DE IMAGÉNES CON OPENCV

*LIBRERÍAS*

```python
import cv2
import matplotlib.pyplot as plt
```

*Obtención de imagen*

```python
img = cv2.imread('C:/Users/FABRICIO/Pictures/fondos/OIP.png')
```

```python
cv2.imshow('imagen',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

*Tamaño de la imagen*

```python
print(img.shape)
```

```python
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)   #invierte colores
plt.subplot(2,1,1)
plt.imshow(img)
plt.subplot(2,1,2)
plt.imshow(rgb_img)
plt.show()
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)   #escala de grises
plt.subplot(2,1,1)
plt.imshow(img)
plt.subplot(2,1,2)
plt.imshow(gray_img)
plt.show()
```

```python
azul, verde, rojo = cv2.split(img)
cv2.imshow('Original',img)
cv2.imshow('Azul',azul)
cv2.imshow('Verde',verde)
cv2.imshow('Rojo',rojo)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
