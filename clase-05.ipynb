# HISTORGRAMA

```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('rice.tif')
imgG=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
hist=cv2.calcHist([img],[0], None, [256], [0,255])
plt.plot(hist)
plt.show()
cv2.imshow('IMG',imgG)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# BINARIZACION

```python
import cv2

img=cv2.imread('rice.tif')
imgG=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
_,imgB=cv2.threshold(imgG,105,255,cv2.THRESH_BINARY) #AQUI SE USA UN VALOR DEL HISTOGRAMA
cv2.imshow('IMG',img)
cv2.imshow('IMG2',imgB)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# CONTEO DE FIGURAS

_DIBUJO DE CONTORNOS_

```python
import cv2
import matplotlib.pyplot as plt
#img=cv2.imread('rice.tif')
imgG=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
_,imgB=cv2.threshold(imgG,105,255,cv2.THRESH_BINARY)
contornos,_=cv2.findContours(imgB.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
cv2.drawContours(img,contornos,-1, (0,255,0),1)
cv2.imshow('IMG',img)
cv2.imshow('IMG2',imgB)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

_CONTEO DE OBJETOS_

```python
import cv2
import matplotlib.pyplot as plt
#img=cv2.imread('rice.tif')
#imgG=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
_,imgB=cv2.threshold(imgG,108,255,cv2.THRESH_BINARY) #AQUI SE USA EL VALOR DEL HISTOGRAMA EN LA SEGUNDA COMA
contornos,_=cv2.findContours(imgB.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
cv2.drawContours(img,contornos,-1, (0,255,0),1)

cont=0
for contour in contornos:
    cont=cont+1
print(cont)

cv2.imshow('IMG',img)
cv2.imshow('IMG2',imgB)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

_ENMARCO_

```python
import cv2
import matplotlib.pyplot as plt
#img=cv2.imread('rice.tif')
imgG=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
_,imgB=cv2.threshold(imgG,105,255,cv2.THRESH_BINARY) #AQUI SE USA EL VALOR DEL HISTOGRAMA EN LA SEGUNDA COMA
contornos,_=cv2.findContours(imgB.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
cv2.drawContours(img,contornos,-1, (0,255,0),1)

cont=0
for contour in contornos:
    area=cv2.contourArea(contour)
    if(area>30):
        cont=cont+1
        x,y,w,h=cv2.boundingRect(contour)
        cv2.rectangle(img,(x,y),(x+w,y+h),(100,255,100),1)
        cv2.putText(img,str(cont),(x,y-5),cv2.FONT_HERSHEY_SIMPLEX,0.4,(100,255,100),1)
    
print(cont)

cv2.imshow('IMG',img)
cv2.imshow('IMG2',imgB)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
