```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('img_py/fruta.bmp')
imgG=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
hist=cv2.calcHist([img],[0], None, [256], [0,255])
plt.plot(hist)
plt.show()
cv2.imshow('IMG',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img=cv2.imread('img_py/fruta.bmp')
imgG=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
_,imgB=cv2.threshold(img,150,255,cv2.THRESH_BINARY) #
cv2.imshow('IMG',img)
cv2.imshow('IMG2',imgB)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
