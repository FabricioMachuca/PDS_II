# Operador Identidad

```python
import cv2
import numpy as np

img = cv2.imread('C:/Users/FABRICIO/Pictures/fondos/186244.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)             #correci√≥n de colores para libreria matplotlib
imgG = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imgO= np.zeros(img.shape,np.uint8)
rows = imgG.shape[0]
cols = imgG.shape[1]

# Recorrido de p√≠xeles
for j in range(rows):
    for i in range(cols):
            imgO[j,i]=imgG[j,i]

# Visualizar la imagen suavizada con filtro
titles = ['Imagen Original', 'Operador Identidad']
images = [img, imgO]

plt.figure(figsize=(10, 5))

for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```

# Operador Inverso

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#img = cv2.imread('C:/Users/FABRICIO/Pictures/fondos/186244.jpg')
imgG = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imgO= np.zeros(img.shape,np.uint8)
rows = imgG.shape[0]
cols = imgG.shape[1]

for j in range(rows):
    for i in range(cols):
        imgO[j,i]=255-imgG[j,i]

# Visualizar la imagen suavizada con filtro
titles = ['Imagen Original', 'Operador Inverso']
images = [img, imgO]

plt.figure(figsize=(10, 5))

for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```

# Operador de Extension

```python
import cv2
import numpy as np

#Valores de escala de p√≠xeles
p1=100
p2=200

#img = cv2.imread('C:/Users/FABRICIO/Pictures/fondos/186244.jpg')
imgG = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imgO= np.zeros(img.shape,np.uint8)
rows = imgG.shape[0]
cols = imgG.shape[1]

for j in range(rows):
    for i in range(cols):
        if imgG[j,i]<=p1 or imgG[j,i]>=p2:
            imgO[j,i]=0
        else:
            num = 255/(p2-p1)
            imgO[j,i]=imgG[j,i]-p1
            prod = [num * i for i in imgO[j,i]] #producto
            imgO[j,i] = prod

titles = ['Imagen Original', 'Operador de Extensi√≥n']
images = [img, imgO]

plt.figure(figsize=(10, 5))

for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```
